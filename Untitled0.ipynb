```python
import numpy as np
import matplotlib.pyplot as plt

# Zufallsdaten erzeugen (reproduzierbar)
np.random.seed(42)
p = 90
t = 0.18
v = 0.2
x = np.linspace(0, 10, p) * t
y1 = .4 + x + np.random.normal(0, v, size=len(x))


# Plot erstellen
fig, ax = plt.subplots(figsize=(6, 4))

ax.plot(x, y1, color='black', linewidth=1)


# Achsen entfernen
ax.set_xticks([])
ax.set_yticks([])

# Gestrichelten Rahmen setzen
for spine in ax.spines.values():
    spine.set_linestyle((0, (5, 5)))  # gestrichelt
    spine.set_linewidth(1.2)
    spine.set_color('black')

# Grenzen etwas anpassen
ax.set_xlim(min(x), max(x))
ax.set_ylim(0, 2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(46)
n =500
dt = 1/252

mu1, sigma1 = 0.5, 0.9


S1 = np.zeros(n)


S1[0] = 100


for t in range(1, n):
    z1 = np.random.normal()
    z2 = np.random.normal()
    S1[t] = S1[t-1] * np.exp((mu1 - 0.5*sigma1**2)*dt + sigma1*np.sqrt(dt)*z1)


fig, ax = plt.subplots(figsize=(6,4))

ax.plot(S1, color='black', linewidth=1)


ax.set_xticks([])
ax.set_yticks([])

for spine in ax.spines.values():
    spine.set_linestyle((0, (5, 5)))
    spine.set_linewidth(1.2)
    spine.set_color('black')

plt.show()
```
